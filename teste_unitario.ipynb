```python
def is_par(n):
    return n % 2 == 0

print("Testes is_par:")
print("is_par(2):", is_par(2))         # True
print("is_par(3):", is_par(3))         # False
print("is_par(0):", is_par(0))         # True
print("is_par(-4):", is_par(-4))       # True
print("is_par(-3):", is_par(-3))       # False
print()

def fatorial(n):
    if n < 0:
        raise ValueError("n deve ser >= 0")
    if n == 0:
        return 1
    resultado = 1
    for i in range(1, n + 1):
        resultado *= i
    return resultado

print("Testes fatorial:")
print("fatorial(0):", fatorial(0))     # 1
print("fatorial(5):", fatorial(5))     # 120
try:
    fatorial(-1)
except ValueError as e:
    print("fatorial(-1): Erro esperado ->", e)
print()
class InsufficientFunds(Exception):
    pass

class Conta:
    def __init__(self):
        self.saldo = 0

    def depositar(self, amount):
        if amount < 0:
            raise ValueError("Valor negativo não permitido")
        self.saldo += amount

    def sacar(self, amount):
        if amount < 0:
            raise ValueError("Valor negativo não permitido")
        if amount > self.saldo:
            raise InsufficientFunds("Saldo insuficiente")
        self.saldo -= amount

print("Testes Conta:")
conta = Conta()
conta.depositar(100)
print("Saldo após depósito de 100:", conta.saldo)  # 100
conta.sacar(50)
print("Saldo após saque de 50:", conta.saldo)      # 50
try:
    conta.sacar(100)
except InsufficientFunds as e:
    print("Saque acima do saldo: Erro esperado ->", e)
try:
    conta.depositar(-2)
except ValueError as e:
    print("Depósito negativo: Erro esperado ->", e)
try:
    conta.sacar(-5)
except ValueError as e:
    print("Saque negativo: Erro esperado ->", e)
print()

import requests
from unittest.mock import patch, Mock

def buscar_clima(cidade):
    response = requests.get(f"https://api.exemplo.com/clima?cidade={cidade}")
    data = response.json()
    if "temperatura" not in data:
        raise ValueError("Resposta inválida: temperatura não encontrada")
    return data["temperatura"]

print("Testes buscar_clima:")

mock_response = Mock()
mock_response.json.return_value = {"temperatura": 25}
with patch('requests.get', return_value=mock_response):
    print("Temperatura em São Paulo:", buscar_clima("São Paulo"))  # 25

mock_response.json.return_value = {}
with patch('requests.get', return_value=mock_response):
    try:
        buscar_clima("CidadeInvalida")
    except ValueError as e:
        print("Erro esperado (sem temperatura):", e)
```
